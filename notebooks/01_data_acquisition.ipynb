# 01 : Acquisition des données

Ce notebook couvre la première étape du projet : le chargement et la découverte 
du dataset principal.

**Source :** Kaggle : Mental Health and Lifestyle Habits 2019-2024  
**Fichier :** `Mental_Health_Lifestyle_Dataset.csv`  
**Objectif :** Charger les données, comprendre leur structure, identifier 
les premières pistes de nettoyage.

```python
# Manipulation des données
import pandas as pd

# Gestion des chemins de fichiers (portable Windows/Mac/Linux)
import os

# On remonte à la racine du projet depuis le dossier notebooks/
# Ainsi tous les chemins relatifs fonctionneront depuis la racine
os.chdir("..")

# Vérification — doit afficher le chemin racine du projet
print(f"Répertoire de travail : {os.getcwd()}")
```

```python
# Chemin relatif vers le fichier CSV (fonctionne depuis la racine du projet)
data_path = os.path.join("data", "raw", "Mental_Health_Lifestyle_Dataset.csv")

# Chargement du CSV dans un DataFrame pandas

df = pd.read_csv(data_path, sep=",")

print(f"Dataset chargé avec succès : {df.shape[0]} lignes, {df.shape[1]} colonnes")
```

```python
# 1. Afficher les 5 premières lignes pour visualiser la structure
print("=== APERÇU DES PREMIÈRES LIGNES ===")
display(df.head())

# 2. Dimensions du dataset
print(f"\n=== DIMENSIONS ===")
print(f"Nombre de lignes    : {df.shape[0]}")
print(f"Nombre de colonnes  : {df.shape[1]}")

#  3. Types de données et valeurs non-nulles par colonne
print("\n=== TYPES DE DONNÉES ===")
print(df.dtypes)
```

```python
# Calcul du nombre de valeurs manquantes par colonne
missing_count = df.isnull().sum()

# Calcul du pourcentage de valeurs manquantes par colonne
missing_percent = (df.isnull().sum() / len(df)) * 100

# On regroupe dans un DataFrame lisible, trié par ordre décroissant
missing_df = pd.DataFrame({
    "Valeurs manquantes" : missing_count,
    "Pourcentage (%)"    : missing_percent.round(2)
}).sort_values("Valeurs manquantes", ascending=False)

print("=== VALEURS MANQUANTES PAR COLONNE ===")
display(missing_df)
```

```python
# Statistiques descriptives pour les colonnes numériques
print("=== STATISTIQUES DESCRIPTIVES ===")
display(df.describe().round(2))
```

```python
# Sélection des colonnes de type "object" (texte/catégories)
categorical_cols = df.select_dtypes(include="object").columns

print("=== VALEURS UNIQUES PAR COLONNE CATÉGORIELLE ===\n")

for col in categorical_cols:
    unique_vals = df[col].unique()
    print(f" {col} ({len(unique_vals)} valeurs uniques) :")
    print(f"   {unique_vals}\n")
```
